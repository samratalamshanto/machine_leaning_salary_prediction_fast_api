```python
def mergeCategories(categoriesDict,cutOff):
  mergeCategoriesDict = {}
  for category,count in categoriesDict.items():
    if count > cutOff:
      mergeCategoriesDict[category] = category
    else:
      mergeCategoriesDict[category] = "Others"
  return mergeCategoriesDict
```

```python
list = df["Country"].value_counts()
list
```

```python
country_map = mergeCategories(df.Country.value_counts(),300)
df["Country"] = df["Country"].map(country_map)  # map() --> list value will be replaced with map's value based on list[i]==map[key]
df.Country.value_counts()
```

```python
fig, ax = plt.subplots(1,1,figsize=(12, 7))

df.boxplot('Salary','Country',ax=ax)
# plt.suptilte('Salary (US$) v Country')
plt.title(' ')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df["YearsCodePro"].unique()
```

```python
def cleanExperience(x):
  if x == "Less than 1 year":
    return 0.5
  elif x == "More than 50 years":
    return 50
  return float(x)
```

```python
df["YearsCodePro"] = df["YearsCodePro"].apply(cleanExperience)
```

```python
df["YearsCodePro"].unique()
```

```python
df.head()
```

```python
df.EdLevel.unique()
```

```python
def cleanEducation(x):
  if x=='Bachelor’s degree (B.A., B.S., B.Eng., etc.)':
    return 'Bachelor’s degree'
  elif (x=='Master’s degree (M.A., M.S., M.Eng., MBA, etc.)'):
    return 'Master’s degree'
  elif (x=='Professional degree (JD, MD, Ph.D, Ed.D, etc.)'):
    return 'Post grad'
  else:
    return 'Less than a Bachelors'

df["EdLevel"] = df["EdLevel"].apply(cleanEducation)
```

```python
df.EdLevel.unique()
```

```python
df = df[df["Country"] != 'Others']
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
edu_encoder = LabelEncoder()
country_encoder = LabelEncoder()
df["EdLevel"] = edu_encoder.fit_transform(df["EdLevel"])
df["Country"] = country_encoder.fit_transform(df["Country"])
```

```python
df["Country"].unique()
```

```python
df["EdLevel"].unique()
```

```python
X = df.drop("Salary",axis=1)
Y = df["Salary"]
```

```python
from sklearn.linear_model import LinearRegression
linnear_reg = LinearRegression()

linnear_reg.fit(X, Y.values)
```

```python
y_pred = linnear_reg.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error = np.sqrt(mean_squared_error(Y, y_pred))
```
